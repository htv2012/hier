# Format Node

```python
import json, pathlib, functools
import xml.etree.ElementTree as ET
```

```python
@functools.singledispatch
def fmt(node) -> str:
    raise NotImplementedError(f"Don't know how to handle type {node.__class__.__name__}: {node!r}")

@fmt.register
def _(node: ET.Element):
    text = (node.text or "").strip()
    attributes = ", ".join(f"{k}={v!r}" for k, v in node.items())
    if attributes:
        attributes = f"  # {attributes}"

    if text:
        out = f"{node.tag}={text!r}{attributes}"
    else:
        out = f"{node.tag}{attributes}"
    return out

@fmt.register
def _(node: str):
    return node

@fmt.register
def _(node: int):
    return f"[{node}]"
```

```python
def try1():
    nodes = [ET.fromstring('<foo x="1" y="2">text for foo</foo>'), "foo", 3]
    for node in nodes:
        print(fmt(node))

try1()
```
